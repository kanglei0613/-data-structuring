```python
shirt_price = 300
pants_price = 350
vest_price = 400

num_shirts = int(input("請輸入上衣的數量？"))
num_pants = int(input("請輸入褲子的數量？"))
num_vests = int(input("請輸入背心的數量？"))

total_amount = (num_shirts * shirt_price) + (num_pants * pants_price) + (num_vests * vest_price)

print(f"訂購服裝的總金額為 {total_amount} 元")
```
