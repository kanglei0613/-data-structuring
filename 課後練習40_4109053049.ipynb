```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")

clarity_order = ["IF", "VVS1", "VVS2", "VS1", "VS2", "SI1", "SI2", "I1"]

sns.boxplot(
    data=diamonds,
    x="clarity",
    y="price",
    order=clarity_order,
    palette="Set3"
)

plt.show()
```
