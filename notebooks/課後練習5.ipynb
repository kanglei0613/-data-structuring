```python
to_do_list = []

for i in range(5):
    work = input(f"請輸入第{i+1}項工作：")
    to_do_list.append(work)

print("\n待辦事項列表:", to_do_list)

first_two = [to_do_list.pop(0), to_do_list.pop(0)]
print("\n取出的工作:", first_two)
print("剩餘的工作:", to_do_list)

last_work = to_do_list.pop()
print("\n取出的工作:", last_work)
print("剩餘的工作:", to_do_list)
```
