```python
def combination(m, n):
    if n == 0 or n == m:
        return 1
    else:
        return combination(m-1, n) + combination(m-1, n-1)

m = int(input("請輸入 m 值？"))
n = int(input("請輸入 n 值？"))

if n > m:
    print("錯誤：n 不可以大於 m")
else:
    result = combination(m, n)
    print(f"從 {m} 取 {n} 個的組合結果為 {result}")
```
