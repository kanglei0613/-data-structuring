```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import LabelEncoder

diamonds = sns.load_dataset("diamonds")

le = LabelEncoder()

for col in ['cut', 'color', 'clarity']:
    diamonds[col + '_encoded'] = le.fit_transform(diamonds[col])

diamonds['cut_encoded'] = le.fit_transform(diamonds['cut'])

print(diamonds[['cut', 'cut_encoded']].head())
```
