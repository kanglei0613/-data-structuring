```python
def check_password_security(password):
    has_digit = any(c.isdigit() for c in password)
    has_alpha = any(c.isalpha() for c in password)
    length = len(password)

    if (has_digit and not has_alpha) or (has_alpha and not has_digit):
        if length < 6:
            return "不安全的密碼"
        else:
            return "不安全的密碼"
    elif has_digit and has_alpha:
        if length < 6:
            return "不安全的密碼"
        elif length < 10:
            return "安全的密碼"
        else:
            return "非常安全的密碼"
    else:
        return "不符合判定條件"

password = input("請輸入密碼：")
print(check_password_security(password))
```
