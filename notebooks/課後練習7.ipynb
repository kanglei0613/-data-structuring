```python
import string

poem1 = "紅豆生南國，春來發幾枝？願君多采擷，此物最相思。"
poem2 = "春眠不覺曉，處處聞啼鳥。夜來風雨聲，花落知多少。"

translator = str.maketrans('', '', string.punctuation)
poem1_clean = poem1.translate(translator)
poem2_clean = poem2.translate(translator)

set1 = set(poem1_clean)
set2 = set(poem2_clean)

common_chars = set1 & set2

print("共同的字:", common_chars)
```
