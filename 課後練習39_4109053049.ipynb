```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamond=sns.load_dataset('diamonds')
```

```python
sns.scatterplot(
    data=diamond,
    x="carat",
    y="price",
    hue="color",
    style="cut",
    palette="light:#5A9"
)

plt.show()
```
